# Starbucks Nutrition EDA Case Study

Dataset: `starbucks_drinkMenu_expanded.csv`

This notebook performs:
1. Loading & inspecting data  
2. Light cleaning  
3. Exploratory Data Analysis  
4. Storytelling conclusions  


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("starbucks_cleaned.csv")
df.head()
```

```python
# Basic structure
print(df.shape)
print(df.dtypes)
print(df.isna().sum())
```

```python
df.describe(include='all').T
```

```python
# Histogram of Calories
df['Calories'].hist(bins=20, figsize=(6,4))
plt.title("Calories Distribution")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sugar vs Calories
plt.scatter(df[' Sugars (g)'], df['Calories'])
plt.title("Sugar vs Calories")
plt.xlabel("Sugars (g)")
plt.ylabel("Calories")
plt.show()
```

```python
df['Sugar_Calories'] = df[' Sugars (g)'] * 4
df.head()
```

```python
df_melted = df.melt(id_vars=['Beverage_category'], value_vars=['Calories', 'Sugar_Calories'], var_name='Metric', value_name='Value')
df_melted.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(16, 8))

# Create a primary axis (ax1)
ax1 = sns.boxplot(x='Beverage_category', y='Value', hue='Metric', data=df_melted, palette={'Calories': 'skyblue', 'Sugar_Calories': 'salmon'})

# Set titles and labels for ax1
ax1.set_title('Distribution of Calories and Sugar Equivalent by Beverage Category', fontsize=16)
ax1.set_xlabel('Beverage Category', fontsize=12)
ax1.set_ylabel('Calories', fontsize=12, color='skyblue')

# Create a secondary axis (ax2) that shares the same x-axis
ax2 = ax1.twinx()

# Set the y-label for ax2. The values are the same as ax1 but this label provides context.
ax2.set_ylabel('Sugar Equivalent (g * 4)', fontsize=12, color='salmon')

# Synchronize y-axis limits to ensure consistent scaling
ax1_ylim = ax1.get_ylim()
ax2.set_ylim(ax1_ylim)

# Rotate x-axis labels for better readability and fix the UserWarning
ax1.set_xticks(ax1.get_xticks()) # Explicitly set ticks before setting tick labels
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right', fontsize=10)

# Get the handles and labels from the artists created by seaborn on ax1
handles, labels = ax1.get_legend_handles_labels()
# Recreate the legend on ax1 with the desired properties
ax1.legend(handles=handles, labels=labels, title='Metric', loc='upper left', bbox_to_anchor=(1.01, 1), borderaxespad=0.)

plt.tight_layout()
plt.show()
```

```python
df.rename(columns={' Total Fat (g)': 'Total Fat (g)'}, inplace=True)
df['Total Fat (g)'] = pd.to_numeric(df['Total Fat (g)'], errors='coerce')
df['Fat_Calories'] = df['Total Fat (g)'] * 9
df.head()
```

```python
df['Pct_Sugar_Calories'] = (df['Sugar_Calories'] / df['Calories'] * 100).fillna(0)
df['Pct_Fat_Calories'] = (df['Fat_Calories'] / df['Calories'] * 100).fillna(0)
df.head()
```

```python
df_agg = df.groupby('Beverage_category')[['Pct_Sugar_Calories', 'Pct_Fat_Calories']].mean().reset_index()
df_agg.head()
```

```python
df_melted_agg = df_agg.melt(id_vars=['Beverage_category'], value_vars=['Pct_Sugar_Calories', 'Pct_Fat_Calories'], var_name='Nutrient_Type', value_name='Percentage')
df_melted_agg.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 8))
ax = sns.barplot(x='Beverage_category', y='Percentage', hue='Nutrient_Type', data=df_melted_agg, palette={'Pct_Sugar_Calories': 'orange', 'Pct_Fat_Calories': 'green'})

plt.title('Average Percentage of Calories from Sugar and Fat by Beverage Category', fontsize=16)
plt.xlabel('Beverage Category', fontsize=12)
plt.ylabel('Average Percentage of Calories', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.legend(title='Nutrient Type', loc='upper left', bbox_to_anchor=(1.01, 1), borderaxespad=0.)

# Add percentage numbers above each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%')

plt.tight_layout()
plt.show()
```

# Executive Narrative: What Drives Calories & Sugar in the Starbucks Menu?

This section reframes our analysis specifically for an **executive audience**.  
The focus is on **product strategy, health positioning, and menu optimization**, not raw code.

---

## 1. Why This Matters

The expanded Starbucks drink menu spans everything from plain brewed coffee to indulgent blended beverages.  
From a business perspective, nutrition is now a **strategic variable**, because it impacts:

- **Brand positioning** (indulgent treats vs. everyday beverages)
- **Regulatory risk** (sugar taxes, labeling requirements)
- **Customer segments** (health-conscious customers vs. treat-seekers)
- **Long-term loyalty** (perception of Starbucks as compatible with a healthy lifestyle)

Our core question:

> **Which parts of the menu disproportionately drive calories and sugar, and where are the best opportunities to improve nutritional profile without harming sales?**

---

## 2. Big Picture: Not All Drinks Are Nutritional Equals

From the EDA, three broad groups emerge:

1. **Very low-calorie “base” coffee drinks**  
   - Examples: brewed coffee, Americanos, some cold brews (without added syrups or cream).  
   - Typically **< 15–20 calories**, minimal sugar.

2. **Moderate-calorie milk-based espresso drinks**  
   - Examples: lattes, cappuccinos, some mochas.  
   - Calories driven by **milk choice** and **number of syrup pumps**.

3. **High-calorie, high-sugar blended and flavored drinks**  
   - Examples: Frappuccinos, heavily flavored seasonal drinks.  
   - Frequently **350–600+ calories**, **very high sugar** per serving.

### Executive takeaway
> The **nutritional “risk”** of the menu is concentrated in a relatively small subset of **blended and heavily flavored beverages**, not in the core coffee offerings.

This means targeted changes can have outsized impact.

---

## 3. Key Insight: Sugar Is the Primary Driver of Calories

When comparing **Calories** to **Sugars (g)**:

- There is a **strong, almost linear relationship** between sugar content and calories.
- Drinks with **> 50 g of sugar** almost always land in the **400+ calorie** range.
- Many of these are **Frappuccinos or dessert-like beverages**.

We also see:

- Brewed coffee and Americanos cluster at **very low calories and sugar**.
- Flavored and blended drinks shift far to the **upper-right** of the sugar–calorie scatterplot.

### Executive takeaway
> If the organization wants to reduce calories in strategic markets or categories, the **most efficient lever is sugar reduction** in a specific subset of drinks.

This can be done via:

- Fewer syrup pumps by default  
- Lower-sugar syrups or reformulations  
- Smaller default sizes in the most calorie-dense lines  

---

## 4. The “Perception Gap”: Drinks That *Feel* Lighter Than They Are

Another important qualitative finding from the EDA:

- Some drinks **appear light or refreshing** (e.g., iced, blended, fruit-forward).
- Yet nutritionally, they perform like **full desserts** when measured by calories and sugar.

This “perception gap” matters because:

- Customers may **underestimate** their intake.
- Over time, this can create **friction** between customers’ health goals and their Starbucks habits.
- It may also attract **regulatory and media scrutiny**, especially where sugar taxes or public-health campaigns are active.

### Executive takeaway
> There is brand value in **aligning perceived lightness with actual nutritional lightness**, especially for iced and fruit-based beverages.

This can be achieved by:

- Creating **truly lighter variants** of popular drinks
- Highlighting **genuinely low-sugar options** in marketing and menu design
- Adjusting **recipe defaults** in the most misleading categories

---

## 5. Opportunities for Action

Based on this exploratory analysis, here are **priority opportunities**:

### 5.1. Targeted Reformulation of Top-Calorie Drinks
Focus on the **highest-calorie, highest-sugar quartile** of drinks:

- Identify the **top ~20% of beverages by calories**.
- For each, explore:
  - Reducing default syrup pumps
  - Offering lighter bases (e.g., sugar-free, lower-fat milk)
  - Adjusting whipped cream or toppings

**Business impact:**  
Large nutritional improvement with **minimal menu complexity**, focusing only on a small set of SKUs.

---

### 5.2. “Everyday Coffee” vs. “Treat” Segmentation

The data naturally divides the menu into:

- **Everyday, low-calorie coffee options**
- **Occasional, high-calorie treats**

Codifying this segmentation could support:

- Clearer **menu labeling and grouping** (e.g., “Light & Daily” vs. “Treat Yourself” sections)
- **Pricing and portion strategies** that encourage:
  - Everyday use of low-calorie drinks
  - Smaller default sizes for dessert-like drinks

**Business impact:**  
Supports both **frequency of visits** (daily drinkers) and **premium pricing** (for indulgent items), while transparently managing nutritional expectations.

---

### 5.3. Health-Conscious Product Line & Marketing

There is an opportunity to lean into:

- Explicitly **low-sugar**, **low-calorie**, or **high-protein** beverages.
- Marketing campaigns that:
  - Highlight these options
  - Provide **simple nutritional cues** (icons, categories)
  - Align with consumer trends toward **wellness** and **functional beverages**

**Business impact:**  
Appeals to health-focused consumers, expands relevance of Starbucks as part of a **sustainable daily routine**, not just an indulgence.

---

## 6. Strategic Summary

From the Starbucks drink nutrition data, the EDA tells a clear strategic story:

1. **Nutritional concern is concentrated** in a subset of high-calorie, high-sugar blended and flavored beverages.  
2. **Sugar content is the dominant driver of calories**, making it the primary lever for nutritional improvement.  
3. There is a **perception gap** for certain iced or blended drinks that seem light but are nutritionally heavy.  
4. With **targeted reformulation, smarter defaults, and clearer menu segmentation**, Starbucks can:
   - Protect and enhance its **brand reputation**
   - Maintain or grow **revenue** from indulgent items
   - Strengthen its position with **health-conscious consumers**

This positions nutrition not as a constraint, but as a **strategic tool** for long-term product and brand evolution.
